* __Задача 1__: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.  
 Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ – 3.

```python
import random
import matplotlib.pyplot as plt
import numpy as np
import time as time

%matplotlib inline

def prod_non_zero_diag(x):
    xi=len(x)-1
    yi=len(x[0])-1
    otv=1
    i=0
    j=0
    while (i<=xi and j<=yi):
        if(x[i][j]):
             otv=otv*x[i][j]
        i+=1
        j+=1
    return otv
    pass
def prod_non_zero_diag_np(x):
    return np.prod(np.diag(x,0)[np.diag(x)!=0])
    pass
def show_graph(time_non_vectorized, time_vectorized, data_size):
    
    f, (ax2) = plt.subplots(1, 1, figsize=(13, 5))
    
    ax2.plot(data_size, time_non_vectorized, label=u"Невекторизованная")
    ax2.plot(data_size, time_vectorized, label=u"Векторизованная")
    ax2.set_title(u"График зависимости времени от меньшей стороны массива")
    ax2.set_xlabel(u"Длина меньшей стороны массива")
    ax2.set_ylabel(u"Время (с)")
    ax2.grid()
    ax2.legend()
    
    f.show()
    pass
time_v=np.zeros(300)
time_n_v=np.zeros(300)
data_size=np.zeros(300)
print(1)
for k in range(300):
    data_size[k]=random.randint(20,800);
data_size.sort()
for k in range(300):
    n=int(data_size[k])
    m=random.randint(n,800+1)
    tim1=time.time();
    a=[[random.randint(-50, 50) for _ in range(m)] for _ in range(n)]
    prod_non_zero_diag(a)
    time_n_v[k]=time.time()-tim1;
    tim1=time.time();
    a1=np.zeros((n,m));
    for i in range(n):
        for j in range(m):
            a1[i][j]=a[i][j]
    prod_non_zero_diag_np(a1)
    time_v[k]=time.time()-tim1;
show_graph(time_n_v,time_v,data_size)
```

  Например, для x = np.array([1, 2, 2, 4]), y = np.array([4, 2, 1, 2]) ответ – True.
  
  


```python
import random
import matplotlib.pyplot as plt
import numpy as np
import time as time

%matplotlib inline

def are_multisets_equal(x, y):
    x.sort()
    y.sort()
    return x==y
    pass
def are_multisets_equal_non(x, y):
    np.sort(x)
    np.sort(y)
    return x==y
    pass
def show_graph(time_non_vectorized, time_vectorized, data_size):
    
    f, (ax2) = plt.subplots(1, 1, figsize=(13, 5))
    
    ax2.plot(data_size, time_non_vectorized, label=u"Невекторизованная")
    ax2.plot(data_size, time_vectorized, label=u"Векторизованная")
    ax2.set_title(u"График зависимости времени от меньшей стороны массива")
    ax2.set_xlabel(u"Длина меньшей стороны массива")
    ax2.set_ylabel(u"Время (с)")
    ax2.grid()
    ax2.legend()
    
    f.show()
    pass
time_v=np.zeros(1000)
time_n_v=np.zeros(1000)
data_size=np.zeros(1000)
print(1)
for k in range(1000):
    data_size[k]=random.randint(20,100000);
data_size.sort()
for k in range(1000):
    n=int(data_size[k])
    tim1=time.time();
    a=[random.randint(-500, 5000000)for _ in range(n)]
    b=[random.randint(-500, 5000000)for _ in range(n)]
    are_multisets_equal(a,b)
    time_n_v[k]=time.time()-tim1;
    tim1=time.time();
    a1=np.zeros(n);
    b1=np.zeros(n);
    for i in range(n):
            a1[i]=a[i]
            b1[i]=b[i]
    are_multisets_equal_non(a1,b1)
    time_v[k]=time.time()-tim1;
show_graph(time_n_v,time_v,data_size)
```

* __Задача 3__: Найти максимальный элемент в векторе x среди элементов, перед которыми стоит нулевой.  
 Например, для x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]) ответ – 5.

```python
import random
import matplotlib.pyplot as plt
import numpy as np
import time as time

%matplotlib inline

def max_after_zero(x):
    max=-100000
    for i in range(len(x)):
        if(x[i]==0 and i+1!=len(x)):
            if(x[i+1]>max):
                max=x[i+1]
    return max
    pass
def max_after_zero_nun(x):
    return np.max(x[1:][np.nonzero(x[:-1] == 0)])
    pass
def show_graph(time_non_vectorized, time_vectorized, data_size):
    
    f, (ax2) = plt.subplots(1, 1, figsize=(13, 5))
    
    ax2.plot(data_size, time_non_vectorized, label=u"Невекторизованная")
    ax2.plot(data_size, time_vectorized, label=u"Векторизованная")
    ax2.set_title(u"График зависимости времени от меньшей стороны массива")
    ax2.set_xlabel(u"Длина меньшей стороны массива")
    ax2.set_ylabel(u"Время (с)")
    ax2.grid()
    ax2.legend()
    
    f.show()
    pass
b=np.array([1,2,3,4]);
time_v=np.zeros(1000)
time_n_v=np.zeros(1000)
data_size=np.zeros(1000)
print(1)
for k in range(1000):
    data_size[k]=random.randint(20,100000);
data_size.sort()
for k in range(1000):
    n=int(data_size[k])
    tim1=time.time();
    a=[random.randint(-500, 500)for _ in range(n)]
    a[0]=0;
    max_after_zero(a)
    time_n_v[k]=time.time()-tim1;
    tim1=time.time();
    a1=np.zeros(n);
    for i in range(n):
            a1[i]=a[i]
    max_after_zero_nun(a1)
    time_v[k]=time.time()-tim1;
show_graph(time_n_v,time_v,data_size)
```

* __Задача 5__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ – (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
import random
import matplotlib.pyplot as plt
import numpy as np
import time as time

%matplotlib inline

def run_length_encoding(x):
    x.sort()
    a=[x[0]]
    b=[]
    prom=1
    for i in range(1,len(x)):
        if(x[i]!=x[i-1]):
            b.append(prom)
            prom=1;
            a.append(x[i])
        else:
            prom+=1;
            
    pass
def run_length_encoding_nun(x):
    dif=np.diff(x)
    h=np.array([-1])
    h=np.append(h, np.nonzero(dif))
    h+=1
    a=x[h]
    h=np.append(h,len(x))
    b=np.diff(h)
    return (a,b)
    pass
def show_graph(time_non_vectorized, time_vectorized, data_size):
    
    f, (ax2) = plt.subplots(1, 1, figsize=(13, 5))
    
    ax2.plot(data_size, time_non_vectorized, label=u"Невекторизованная")
    ax2.plot(data_size, time_vectorized, label=u"Векторизованная")
    ax2.set_title(u"График зависимости времени от меньшей стороны массива")
    ax2.set_xlabel(u"Длина меньшей стороны массива")
    ax2.set_ylabel(u"Время (с)")
    ax2.grid()
    ax2.legend()
    
    f.show()
    pass
b=np.array([1,2,3,4]);
time_v=np.zeros(1000)
time_n_v=np.zeros(1000)
data_size=np.zeros(1000)
print(1)
for k in range(1000):
    data_size[k]=random.randint(20,100000);
data_size.sort()
for k in range(1000):
    n=int(data_size[k])
    tim1=time.time();
    a=[random.randint(-50000, 50000)for _ in range(n)]
    run_length_encoding(a)
    time_n_v[k]=time.time()-tim1;
    tim1=time.time();
    a1=np.zeros(n);
    for i in range(n):
            a1[i]=a[i]
    run_length_encoding_nun(a1)
    time_v[k]=time.time()-tim1;
show_graph(time_n_v,time_v,data_size)
```

```python
import time
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
def pairwise_distance(x, y):
    a1=[[0 for _ in range(len(y))] for _ in range(len(x))]
    for i in range(len(x)):
        for j in range(len(y)):
            a1[i][j]=((x[i][0]-y[i][0])**2+(x[i][1]-y[i][1])**2)**0.5;
    return a1
    pass
def pairwise_distance_nun(x, y):
    a1=np.arange(1,len(x)*len(y)+1)
    a1 = np.reshape(a1,(len(x),len(y)))
    a1[np.nonzero(a1)]=((x[np.nonzero(a1)[0]][0]-y[np.nonzero(a1)[1]][0])**2+(x[np.nonzero(a1)[0]][1]-y[np.nonzero(a1)[1]][1])**2)**0.5
    return a1
    pass
def show_graph(time_non_vectorized, time_vectorized, data_size):
    
    f, (ax2) = plt.subplots(1, 1, figsize=(13, 5))
    
    ax2.plot(data_size, time_non_vectorized, label=u"Невекторизованная")
    ax2.plot(data_size, time_vectorized, label=u"Векторизованная")
    ax2.set_title(u"График зависимости времени от меньшей стороны массива")
    ax2.set_xlabel(u"Длина меньшей стороны массива")
    ax2.set_ylabel(u"Время (с)")
    ax2.grid()
    ax2.legend()
    
    f.show()
    pass
time_v=np.zeros(300)
time_n_v=np.zeros(300)
data_size=np.zeros(300)
for k in range(300):
    data_size[k]=random.randint(20,1000);
data_size.sort()
for k in range(300):
    n=int(data_size[k])
    tim1=time.time();
    a=[[random.randint(-50000, 50000) for _ in range(2)] for _ in range(n)]
    b=[[random.randint(-50000, 50000) for _ in range(2)] for _ in range(n)]
    pairwise_distance(a,b)
    time_n_v[k]=time.time()-tim1;
    tim1=time.time();
    a1=np.zeros((n,2));
    b1=np.zeros((n,2));
    for i in range(n):
            a1[i][0]=a[i][0]
            a1[i][1]=a[i][1]
            b1[i][0]=b[i][0]
            b1[i][1]=b[i][1]
    pairwise_distance_nun(a1,b1)
    time_v[k]=time.time()-tim1;
show_graph(time_n_v,time_v,data_size)
```
